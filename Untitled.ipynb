```python
import matplotlib.pyplot as plt
import scipy.signal as sig
import numpy as np
from math import pi
#from scipy.signal import firwin, freqz, sig
```

```python
N = 8 #tap
Fs = 1000 #sampling frequency (Hz)
fc = 50 #cutoff frequency
w_c = 2*fc/Fs #normalize frequency
t = sig.firwin(N,w_c)
[w,h] = sig.freqz(t, worN=2000) #frequency response 

w = Fs*w/(2*pi) #convert to Hz
h_db = 20*np.log10(abs(h))


plt.figure(figsize = (10,6))
plt.plot(w,h_db)
plt.vlines(50, ymin = min(h_db), ymax = 0, colors = 'red', linestyle = '--', label = 'Cutoff frequency = 50 Hz')
plt.legend()
plt.title('Fir filter Response')
plt.xlabel('Frequency(Hz)')
plt.ylabel('Amplitude(dB)')
plt.grid('on')
```
