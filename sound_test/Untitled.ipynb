```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
import wave
```

```python
input_file = np.loadtxt('input_sound.txt')
output_file = np.loadtxt('output_sound.txt')
```

```python
fig = plt.figure()
plt.plot(input_file)
plt.plot(output_file)
```

```python
input_fft = np.fft.fft(input_file)
output_fft = np.fft.fft(output_file)
freq = np.fft.fftfreq(len(output_file), 1/44100)
```

```python
fig, [ax1, ax2] = plt.subplots(figsize = (15,5), nrows=1, ncols=2)

fig.suptitle('Spectrum analysis of Star Wars soundtrack')

ax1.plot(np.abs(freq), np.abs(input_fft[:-2]), color = 'lightblue', label = 'Input')
ax1.plot(np.abs(freq), np.abs(output_fft), label = 'FPGA output')
ax1.vlines(551.25, ymin = 0, ymax = 140000, zorder=3, colors = 'red', linestyle = '--', label = 'Cutoff frequency = 551.25 Hz')
#ax.set_yscale('log')
ax1.grid()
ax1.set_ylabel('Amplitude')
ax1.set_xlabel('Frequency (Hz)')
ax1.legend()

ax2.plot(np.abs(freq), np.abs(input_fft[:-2]), color = 'lightblue', label = 'Input')
ax2.plot(np.abs(freq), np.abs(output_fft), label = 'FPGA output')
ax2.vlines(551.25, ymin = 0, ymax = 140000, zorder=3, colors = 'red', linestyle = '--', label = 'Cutoff frequency = 551.25 Hz')
ax2.set_yscale('log')
ax2.set_ylabel('Amplitude')
ax2.set_xlabel('Frequency (Hz)')
ax2.grid()
ax2.legend()
```

```python
fig.savefig('spectrum_sw.pdf')
```
